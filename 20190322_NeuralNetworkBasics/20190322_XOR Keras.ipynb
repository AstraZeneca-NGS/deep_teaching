The following is a Keras implementation to solve the XOR problem. Here we try to use the same exact conditions that were used in our manually coded example, namely
* The same input and output data
* Learning rate (`eta` in manual example, `lr` in this notebook) = 1
* Random weight initialization between `-0.5` and `0.5`
* Sigmoid activation functions
* Mean squared error loss function (see `def backprop(self, num_in):` in the `Network` class in our manual example)
* The same training regime, i.e. using all 4 input datapoints (`batch_size=4`) for each training iteration (total 2000 iterations i.e. `epochs`)

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.initializers import RandomUniform
from keras.losses import mean_squared_error
from keras import optimizers
import numpy as np
import matplotlib.pyplot as plt
```

```python
RAND_WEIGHT_FACTOR = 0.5

XOR_INPUTS = np.array([[-1, -1], [-1, 1], [1, -1], [1, 1]])
XOR_OUTPUTS = np.array([0.1, 0.9, 0.9, 0.1])
```

```python
# specify how we're to initialize our neuron weights
ru = RandomUniform(minval=-1*RAND_WEIGHT_FACTOR, maxval=1*RAND_WEIGHT_FACTOR)
```

```python
def makeNN(): # a function we can call to build our model
    myNN = Sequential() # instantiate a model
    
    # note: Dense() default is to initialize biases to 0
    myNN.add(Dense(2,activation='sigmoid',kernel_initializer=ru,input_shape=(XOR_INPUTS.shape[1],),bias_initializer='ones'))
    myNN.add(Dense(1,activation='sigmoid',kernel_initializer=ru,bias_initializer='ones'))
    sgd = optimizers.SGD(lr=1)
    myNN.compile(loss=mean_squared_error,optimizer=sgd)
    return myNN
```

```python
# make our model
myNN = makeNN() 

# train our model
history = myNN.fit(x=XOR_INPUTS,y=XOR_OUTPUTS,batch_size=4,epochs=2000)
```

```python
# plot model loss with training
plt.plot(history.history['loss'])
```

```python
# evaluate our model
myNN.predict(XOR_INPUTS)
```
